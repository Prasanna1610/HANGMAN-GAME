```python
import random
fruits = ("Apple", "Banana", "Mango", "Blueberry", "Pineapple","Grape", "Kiwi", "Orange","Raspberry")
print(random.sample(fruits, 5))
fruit = random.choice(fruits).upper()
```

```python
j=0
k=0
print("The length of the word is ",len(fruit))
guesses=[]
for i in range(len(HANGMAN)-1):
    fruity=list(fruit)
    user=input("enter a letter: ")
    if user in guesses:
        print("you already guessed this letter,type another letter")
    elif user in fruit:
        guesses.append(user)
        cont=fruit.count(user)
        k=k+cont
        print("you guessed",k,"letters")
    else:
        print("the letter is not present")
    j=j+1
    print(HANGMAN[j])
    if k==len(fruit):
        print("you guessed the word,congratulations!!")
        print("the word is",fruit)
        break
    if j==len(HANGMAN)-1:
        print("you lost")
        print("the word is ",fruit)
        break

```
